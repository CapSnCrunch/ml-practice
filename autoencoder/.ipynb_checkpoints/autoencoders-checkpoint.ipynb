# MNIST

```python
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from torch.utils.data import DataLoader

transform = transforms.Compose([transforms.ToTensor()])
mnist_trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)

batch_size = 100
train_loader = DataLoader(dataset=mnist_trainset, batch_size=batch_size, shuffle=True)
```

# Autoencoder

```python
import torch
from torch import nn

class AutoEncoder(nn.Module):
    def __init__(self, input_dims, latent_dims):
        super().__init__()

        self.latent_dims = latent_dims

        self.encoder = nn.Sequential(
            nn.Linear(input_dims, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, latent_dims)
        )

        self.decoder = nn.Sequential(
            nn.Linear(latent_dims, 32),
            nn.ReLU(),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, input_dims)
        )

    def encode(self, x):
        with torch.no_grad():
            return self.encoder(x)

    def decode(self, x):
        with torch.no_grad():
            return self.decoder(x)
    
    def forward(self, x):
        encoding = self.encoder(x)
        decoding = self.decoder(encoding)
        return decoding
```

```python
models = []

for dim in [2, 8]:
    print(f'Training model with {dim} latent dimensions')
    
    image = np.array(mnist_trainset[0][0])
    model = AutoEncoder(image.size, dim)
    # model.to(mps_device)
    
    loss_fn = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, betas=(0.9, 0.999))
    
    model.train()

    for epoch in range(5):
        print(f'Epoch {epoch + 1}')
        for batch_idx, (x, _) in enumerate(train_loader):
            x = x.view(batch_size, image.size)
            x = x.to(model.parameters().__next__().device)
            
            pred = model(x)
            loss = loss_fn(pred, x)
        
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
    
    models.append(model)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(len(models) + 1, 14, figsize=(14, len(models) + 1))

for j in range(len(axs[0])):
    index = np.random.randint(len(mnist_trainset), size=1)[0]
    image = mnist_trainset[index][0]
    axs[0, j].imshow(image[0], cmap='gray')
    axs[0, j].axis('off')
    
    for i, model in enumerate(models):
        model.eval()
        compressed_image = model.forward(image.flatten()).reshape((28,28))
        axs[i+1, j].imshow(compressed_image.detach().numpy(), cmap='gray')
        axs[i+1, j].axis('off')
```

```python
from sklearn.manifold import TSNE
from matplotlib.lines import Line2D

fig, axs = plt.subplots(1, len(models), figsize=(6, 12))
cmap = plt.get_cmap('tab10')

for index, model in enumerate(models):
    
    points = []
    labels = []
    
    for i in range(5000):
        image, label = mnist_trainset[i]
        encoded = models[index].encode(torch.tensor(np.array(image)).flatten())
        points.append(encoded.detach().numpy())
        labels.append(label)
    
    points = np.array(points)
    labels = np.array(labels)

    tsne = TSNE(n_components=2)
    points_tsne = tsne.fit_transform(points)

    for label in np.unique(labels):
        label_points = points_tsne[labels == label]
        axs[index].scatter(label_points[:, 0], label_points[:, 1], color=cmap(label), label=f'{label}', s=1)
        axs[index].axis('off')
        
    legend_elements = [
        Line2D([0], [0], marker='o', color='w', label=f'{i}', markerfacecolor=cmap(i), markersize=10) for i in range(10)
    ]

    axs[index].set_title(f'MNIST Encoding with {model.latent_dims} Latent Dimensions')
    axs[index].legend(handles=legend_elements, title="Digits", loc="upper right", frameon=True, fontsize='medium')
```

# Variational Autoencoder

```python
import torch
from torch import nn

class VariationalAutoEncoder(nn.Module):
    def __init__(self, input_dims, latent_dims):
        super().__init__()

        self.latent_dims = latent_dims

        self.encoder = nn.Sequential(
            nn.Linear(input_dims, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, latent_dims)
        )

        self.mean_layer = nn.Linear(latent_dims, latent_dims)
        self.logvar_layer = nn.Linear(latent_dims, latent_dims)

        self.decoder = nn.Sequential(
            nn.Linear(latent_dims, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, input_dims),
            nn.Sigmoid()
        )

    def encode(self, x):
        with torch.no_grad():
            x = self.encoder(x)
            mean, logvar = self.mean_layer(x), self.logvar_layer(x)
            return mean, logvar

    def decode(self, x):
        with torch.no_grad():
            return self.decoder(x)
    
    def reparameterize(self, mean, logvar):
        std = torch.exp(0.5 * logvar)
        epsilon = torch.randn_like(std)
        z = mean + std * epsilon
        return z

    def forward(self, x):
        x = self.encoder(x)
        mean, logvar = self.mean_layer(x), self.logvar_layer(x)
        
        z = self.reparameterize(mean, logvar)
        x_hat = self.decoder(z)
        
        return x_hat, mean, logvar
```

```python
# KL Divergence
def kl_loss_function(x, x_hat, mean, logvar, beta=0.5, print_error=False):
    # reproduction_loss = ((x - x_hat)**2).sum()
    reproduction_loss = nn.functional.binary_cross_entropy(x_hat, x, reduction='sum')
    KLD = - 0.5 * torch.sum(1 + logvar - mean.pow(2) - logvar.exp())
    if print_error:
        print(reproduction_loss, KLD, (1 - beta) * reproduction_loss + beta * KLD)
    return (1 - beta) * reproduction_loss + beta * KLD
```

```python
vae_models = []

for dim in [2, 8]:
    print(f'Training model with {dim} latent dimensions')
    
    image = np.array(mnist_trainset[0][0])
    model = VariationalAutoEncoder(image.size, dim)
    
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, betas=(0.9, 0.999))
    
    model.train()

    for epoch in range(15):
        # print(f'Epoch {epoch + 1}')
        for batch_idx, (x, _) in enumerate(train_loader):
            x = x.view(batch_size, image.size)
            x = x.to(model.parameters().__next__().device)

            optimizer.zero_grad()

            x_hat, mean, logvar = model(x)
            loss = kl_loss_function(x, x_hat, mean, logvar, 0.5)
            
            loss.backward()
            optimizer.step()
    
    vae_models.append(model)
```

```python
def generate_digit(mean, var, model):
    z_sample = torch.tensor([[mean, var]], dtype=torch.float)
    x_decoded = model.decode(z_sample)
    digit = x_decoded.detach().cpu().reshape(28, 28) # reshape vector to 2d array
    plt.imshow(digit, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
def plot_latent_space(model, xvar=0, yvar=1, scale=1.0, n=25, digit_size=28, figsize=15):
    # display a n*n 2D manifold of digits
    figure = np.zeros((digit_size * n, digit_size * n))

    # construct a grid 
    grid_x = np.linspace(-scale, scale, n)
    grid_y = np.linspace(-scale, scale, n)[::-1]

    for i, yi in enumerate(grid_y):
        for j, xi in enumerate(grid_x):
            mean = [0] * model.latent_dims
            mean[xvar] = xi
            mean[yvar] = yi
            
            z_sample = torch.tensor([mean], dtype=torch.float)
            x_decoded = model.decode(z_sample)
            digit = x_decoded[0].detach().cpu().reshape(digit_size, digit_size)
            figure[i * digit_size : (i + 1) * digit_size, j * digit_size : (j + 1) * digit_size,] = digit

    plt.figure(figsize=(figsize, figsize))
    plt.title('VAE Latent Space Visualization')
    start_range = digit_size // 2
    end_range = n * digit_size + start_range
    pixel_range = np.arange(start_range, end_range, digit_size)
    sample_range_x = np.round(grid_x, 1)
    sample_range_y = np.round(grid_y, 1)
    plt.xticks(pixel_range, sample_range_x)
    plt.yticks(pixel_range, sample_range_y)
    plt.xlabel("mean, z [0]")
    plt.ylabel("var, z [1]")
    plt.imshow(figure, cmap="Greys_r")
    plt.show()


plot_latent_space(vae_models[1], xvar=1, yvar=2, scale=5, digit_size=28)
```
